# Checking for the NaN values and finding ways to fill them

```python
df.isna().sum()
```

```python
# checking for unique values in the column
unique = {col: df[col].unique() for col in df.columns}
unique

# this is not feasible here
```

```python
'''
for features like isWide, isNoBall,Byes,LegByes,Penalty 
good way to NaN values would be 0
AND

for features with dtype='object' like dismissal_kind:
1. One approach is to first get the unique values of dismissal kind and then using the most frequent dismissal kind, we can replace NaN values with it.

for player_dismissed feature we can just put 0 in place of NaN since no player is dissmissed
'''

column_with_float_dtype = ['isWide','isNoBall','Byes','LegByes','Penalty']
for c in column_with_float_dtype:
    df[c] = df[c].fillna(0)
```

```python
df.isna().sum()
```

```python
# now cleaning column_with_float_dtype
# most common way of dissmisal
# print(list(df['dismissal_kind'].mode())) # caught
# but this is not feasible as this would give caught on every NaN....so instead we will use Not Out
df['dismissal_kind'] = df['dismissal_kind'].fillna("Not Out")
```

```python
df.isna().sum()
```

```python
# now cleaning last column: player_dismissed with "0"
df['player_dismissed'] = df['player_dismissed'].fillna("None Dismissed")
```

# Summary from the dataset

```python
# starting from number of matches per season,
# introducing a new column names year and then grouping with that column
df['year'] = pd.to_datetime(df['date']).dt.year
```
