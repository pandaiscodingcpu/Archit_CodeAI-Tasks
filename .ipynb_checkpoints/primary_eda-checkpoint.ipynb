# Checking for the NaN values and finding ways to fill them

```python
df.isna().sum()
```

```python
# checking for unique values in the column
unique = {col: df[col].unique() for col in df.columns}
unique

# this is not feasible here
```

```python
'''
for features like isWide, isNoBall,Byes,LegByes,Penalty 
good way to NaN values would be 0
AND

for features with dtype='object' like dismissal_kind:
1. One approach is to first get the unique values of dismissal kind and then using the most frequent dismissal kind, we can replace NaN values with it.

for player_dismissed feature we can just put 0 in place of NaN since no player is dissmissed
'''

column_with_float_dtype = ['isWide','isNoBall','Byes','LegByes','Penalty']
for c in column_with_float_dtype:
    df[c] = df[c].fillna(0)
```

```python
df.isna().sum()
```

```python
# now cleaning column_with_float_dtype
# most common way of dissmisal
# print(list(df['dismissal_kind'].mode())) # caught
# but this is not feasible as this would give caught on every NaN....so instead we will use Not Out
df['dismissal_kind'] = df['dismissal_kind'].fillna("Not Out")
```

```python
df.isna().sum()
```

```python
# now cleaning last column: player_dismissed with "0"
df['player_dismissed'] = df['player_dismissed'].fillna("None Dismissed")
```

# Summary from the dataset

**In this we are going to perform practices such as number of matches per season, runs scored per season winner of each season etc**

```python
# starting from number of matches per season,
# introducing a new column names year and then grouping with that column
df['year'] = pd.to_datetime(df['date']).dt.year
```

```python
matches_per_year = df.groupby('year')['matchId'].nunique()
matches_per_year
```

```python
plt.xlabel("Year")
plt.ylabel("Number of matches")
plt.xticks(matches_per_year.index)
plt.xticks(rotation=45)
plt.xlim(2007,2025)
plt.ylim(0,100)
plt.grid(True)
plt.plot(matches_per_year.index,matches_per_year.values,linestyle="--",marker="o",color="b",label="Matches per year")
plt.legend()
plt.title("Number of Matches per year (from 2008 to 2024)")
plt.text(2008,40,f"Highest number of matches were played in the year: {matches_per_year.idxmax()}",color="red")
plt.text(2008,30,f"Lowest number of matches were played in the year: {matches_per_year.idxmin()}",color="green")
plt.text(2008,20,f"Average number of matches were played in the year: {int(matches_per_year.values.mean())}",color="blue")
plt.show()

# idxmax() --> returns a Series with the index of the maximum value for each column.
```

## Runs Scored per season

```python
df['total_runs'] = df['batsman_runs']+df['extras']
```

```python
runs_per_season = df.groupby('year')['total_runs'].sum()
```

```python
runs_per_season
```

```python
plt.xlabel("Year")
plt.ylabel("Number of runs")
plt.xticks(runs_per_season.index)
plt.xticks(rotation=45)
plt.xlim(2007,2025)
plt.ylim(15000,30000)
plt.grid(True)
plt.plot(runs_per_season.index,runs_per_season.values,linestyle="--",marker="o",color="g",label="Runs per season")
plt.legend()
plt.title("Runs per year (from 2008 to 2024)")
plt.text(2008,28000,f"Highest Runs were in the year: {runs_per_season.idxmax()}",color="red")
plt.text(2008,26000,f"Lowest Runs were in the year: {runs_per_season.idxmin()}",color="green")
plt.text(2008,24000,f"Average runs per season: {int(runs_per_season.values.mean())}",color="blue")
plt.show()
```

# Most runs scored in a single year

```python
'''
Approach:
1. We can't use total_runs column, so instead we will use innings wise score per year.
2. And then the highest score would be displayed
'''
# first we need to calculate runs per match
runs_per_match = df.groupby('matchId')['total_runs'].sum()
```

```python
runs_per_match
```

```python
# runs_per_match_per_innings
runs_per_match_per_innings = df.groupby(['matchId','inning','year','batting_team'])['total_runs'].sum()
```

```python
runs_per_match_per_innings
```

```python
# print(type(runs_per_match_per_innings.index)) <class 'pandas.core.indexes.multi.MultiIndex'>
idx = runs_per_match_per_innings.groupby(level='year').idxmax()
```

```python
idx
```

```python
max_runs_per_year = runs_per_match_per_innings.loc[idx].reset_index()
```

```python
max_runs_per_year
```

```python
# now we have to visualize this data using columns: year,batting_team,total_runs
# basically the plot will be only year vs total_runs but in that we will include the teams which socred that runs
# Plot
plt.figure(figsize=(10,6))
plt.xlabel("Year")
plt.ylabel("Highest score in a year")
plt.xticks(max_runs_per_year['year'])
plt.xticks(rotation=45)
plt.xlim(2007, 2025)
plt.ylim(200, 300)
plt.grid(True)
plt.plot(max_runs_per_year['year'], max_runs_per_year['total_runs'],
         linestyle="--", marker="o", color="g", label="Highest Runs per season")
max_idx = max_runs_per_year['total_runs'].idxmax()
plt.legend()
plt.title("Highest Runs per year (from 2008 to 2024)")
plt.text(2008, 295, f"Highest: {max_runs_per_year.loc[max_idx, 'total_runs']} in {max_runs_per_year.loc[max_idx, 'year']}",
         color="red")
plt.text(2008, 285, f"Lowest: {max_runs_per_year.loc[max_runs_per_year['total_runs'].idxmin(), 'total_runs']} in {max_runs_per_year.loc[max_runs_per_year['total_runs'].idxmin(), 'year']}",
         color="green")

plt.show()
```

## TOP BATTING PERFORMERS

```python
batsmen_performance = df.groupby(['batsman','batting_team'])['batsman_runs'].sum()
```

```python
batsmen_performance
```

```python
batsmen_performance.idxmin()
```

```python
batsmen_performance.idxmax()
```

```python
# Displaying top 10 run scorers of all time in IPL since 2008
top_10_batsmen = batsmen_performance.sort_values(ascending=False).head(10)
```

```python
top_10_batsmen
```

```python
top_10_batsmen = top_10_batsmen.reset_index() # resetting index positions
top_10_batsmen.columns = ['Batsman', 'Batting Team', 'Total Runs'] # defining the colun

print(top_10_batsmen)
```

```python
# plotting a horizontal bar graph for the above data
plt.barh(top_10_batsmen['Batsman'] + ' (' + top_10_batsmen['Batting Team'] + ')',
         top_10_batsmen['Total Runs'],
         color='skyblue')
plt.xlabel("Total Runs")
plt.ylabel("Batsman (Team)")
plt.title("Top 10 Run Scorers in IPL since 2008")
plt.gca().invert_yaxis()  # highest at top
plt.show()
```
